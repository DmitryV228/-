1.Последний релиз исполнителя, чья песня на 1-м месте в чарте

```python
cursor.execute("""SELECT latest_release FROM artists WHERE musician =
               (SELECT musician FROM song WHERE songid =
                (SELECT songid FROM chart WHERE place_on_the_chart = '1'))""")
result = cursor.fetchone()
print(result[0])
```

4. Cписок альбомов типа сингл в чарте

5. Cписок песен, которые идут дольше 2-х минут

```python
cursor.execute("""SELECT song_name FROM chart WHERE songid in (SELECT songid FROM song WHERE duration > '00:02:00')""")
result = cursor.fetchall()
print(result)
```
